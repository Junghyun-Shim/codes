```python

df.columns=['Date','Price','Open','High','Low','Vol','Change']

df['Price'] =df.Price.str.replace(',', '')
df['Open'] =df.Open.str.replace(',', '')
df['High'] =df.High.str.replace(',', '')
df['Low'] =df.Low.str.replace(',', '')
df['Date'] =df.Date.str.replace('년', '-')
df['Date'] =df.Date.str.replace('월', '-')
df['Date'] =df.Date.str.replace('일', '')
#df['Date']=pd.to_datetime(df['Date']) #유형을 datetime으로 변경

df
```

```python
  
for i in range(df.shape[0]):
    a = df.iloc[i]['Vol']

#         if a.find("-") >0:
#             a= df['Vol'][i].replace('-',"0")

    if a.find("M") > 0:
        a = a.rstrip('M')
        a = float(a)*1000000
        df.iloc[i]['Vol'] = a

    elif a.find("K") > 0:
        a = a.rstrip('K')
        a = float(a)*1000
        df.iloc[i]['Vol'] = a

    
            
df.to_csv("./변경후.csv",index=False)
```
