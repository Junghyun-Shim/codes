# 단위 분리

```python
import pandas as pd
```

```python
# 테스트용 데이터 불러오기
# df = pd.read_excel("./data.xlsx")
df = pd.read_csv("./S&P_500.csv")
df
```

```python
# 위에 테스트 조합해서 최종 결과물 
# 데이터 불러오고나서 요거만 돌리면 됨

for i in range(df.shape[0]):
    a = df.iloc[i]['Vol.']
    if a.find("M") > 0:
        a = a.rstrip('M')
        a = float(a)*1000000
        df.iloc[i]['Vol.'] = a
        
    elif a.find("K") > 0:
        a = a.rstrip('K')
        a = float(a)*1000
        df.iloc[i]['Vol.'] = a
        

# result_save = df.to_excel("./data2.xlsx")
result_save = df.to_csv("./S&P_500_2.csv")
```
