```python
x= df[df.columns.difference(["최종 번호", "직전학기평균학점"])]
x
```

```python
y = df["직전학기평균학점"]
y.head()
```

```python
y_list = list(y.values)
type(y_list)
```

```python
len(y_list)
```

```python
for i in range(17268):
    if y_list[i] >= 4.0:
        y_list[i] = "A"
    elif y_list[i] >= 3.0:
        y_list[i] = "B"
    elif y_list[i] >= 2.0:
        y_list[i] = "C"
    else:
        y_list[i] = "D"
y_list
```

```python
y = pd.Series(y_list)
```

```python
type(y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc.fit(x_train)
```

```python
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
```

```python
df_tree = DecisionTreeClassifier(criterion = 'entropy', max_depth=3, random_state=0)
df_tree.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred_tr = df_tree.predict(x_test)
print("Accu : %.2f" % accuracy_score(y_test, y_pred_tr))
```
